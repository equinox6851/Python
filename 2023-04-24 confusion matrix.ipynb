```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(font_scale=2)
```

### 1. Basic Confusion Matrix

```python
array = [[5, 0, 0, 0],
        [0, 10, 0, 0],
        [0, 0, 15, 0],
        [0, 0, 0, 5]]

df_cm = pd.DataFrame(array, index=[i for i in "ABCD"],
                    columns=[i for i in "ABCD"])
plt.figure(figsize = (10,7))
plt.title("confusion matrix without confusion")
sns.heatmap(df_cm, annot=True)
```

### 2. Confusion Matrix with confusion

```python
array = [[9, 1, 0, 0], # A가 들어와서 9개는 맞추고 1개는 B라고 했다.
        [1, 15, 3, 1], # B가 들어와서 15개는 맞추고 1개는 A, 3개는 C, 하나는 D라고 했다.
        [5, 0, 24, 1], # C가 들어와서 24개는 맞추고 5개는 A, 1개는 D라고 했다.
        [0, 4, 1, 15]] # D가 들어와서 15개는 맞추고 4개는 B, 1개는 C라고 했다.

df_cm = pd.DataFrame(array, index=[i for i in "ABCD"],
                    columns=[i for i in "ABCD"])
plt.figure(figsize = (10,7))
plt.title("confusion matrix without normalization")
sns.heatmap(df_cm, annot=True)
```

### 3. Confusion Matrix with normalization

```python
import numpy as np
from numpy import linalg as LA
```
